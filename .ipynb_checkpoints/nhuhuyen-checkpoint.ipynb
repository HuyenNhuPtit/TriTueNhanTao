```python
import pandas as pd
import numpy as np
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from pyvi import ViTokenizer
from string import punctuation
```

```python
vectorizer = CountVectorizer()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
negative_data = pd.read_csv('negative_data.csv')
positive_data = pd.read_csv('positive_data.csv')
```

```python
negative_data.info()
```

```python
positive_data.info()
```

```python
data = negative_data.append(positive_data, ignore_index=True)
```

```python
data.to_csv('huyen_lananh.csv', index = False)
```

## Bat dau tien xu ly du lieu

```python
# xay dung kho stop word
stop_word = []
with open("stopword.txt",encoding="utf-8") as f :
  t = f.read()
  for word in t.split() :
      stop_word.append(word)
  f.close()
punc = list(punctuation)
stop_word = stop_word + punc
```

```python
def pre_processing(text):
    text_lower = text.lower()
    text_token = ViTokenizer.tokenize(text_lower)
    
    return text_token
```

```python
data['Review'] = data['Review'].apply(pre_processing)
```

```python
data
```

```python
X = vectorizer.fit_transform(data['Review'])
```

```python
y = data['Label']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size = 0.2)
```

```python
# xay dung mo hinh hoc may
lr = LogisticRegression().fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred, y_test)
```
